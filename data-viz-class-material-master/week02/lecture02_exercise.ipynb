## Setup — Run this cell first


```python
import pandas as pd
import numpy as np

# World Happiness Repor|t 2023 — representative data
# Source: https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2023

df = pd.read_csv('../data/world_happiness_2023.csv')
df.columns = ['Country','Region','Happiness_Score','GDP','Social_Support',
              'Life_Expectancy','Freedom','Generosity','Corruption']


print(f"Dataset: {len(df)} countries, {len(df.columns)} columns")
print(df.head())
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Explore the dataset before you start
print("Regions in dataset:")
print(df['Region'].value_counts())
print("\nScore range:", df['Happiness_Score'].min(), "–", df['Happiness_Score'].max())
print("\nBottom 10 countries:")
print(df.nsmallest(10, 'Happiness_Score')[['Country','Region','Happiness_Score']])
```

## Task 1 — Regional Comparison Bar Chart

**What to build:** A horizontal bar chart showing the **average happiness score by region**, sorted from highest to lowest.

**Requirements:**
- Horizontal orientation (category names are long)
- Sorted by score, descending (so the happiest region is at the top)
- Zero baseline on x-axis
- At least one design choice that goes beyond the Plotly default (colour, annotation, labels, etc.)
- An insight title that answers: *which region stands out and why does it matter?*

**Hint:** Use `df.groupby('Region')['Happiness_Score'].mean()` to compute the averages.


```python
# Task 1: Regional comparison bar chart
# -------------------------------------

# Step 1: Compute average happiness score by region
region_avg = (df.groupby('Region')['Happiness_Score']
              .mean()
              .reset_index()
              .sort_values('Happiness_Score'))  # sort for horizontal bar

print(region_avg)

# Step 2: Build your chart
# YOUR CODE HERE

fig = px.bar(
    region_avg,
    x='Happiness_Score',
    y='Region',
    orientation='h',
    color='Happiness_Score',
    color_continuous_scale=['#d73027', '#fee08b', '#1a9850'],
    text=region_avg['Happiness_Score'].round(2),
)

fig.update_traces(
    textposition='outside',
    textfont_size=11,
)

fig.update_layout(
    title=dict(
        text='Western Europe and North America Lead Global Happiness<br>'
             '<sup>Other regions trail by more than 2 points on average</sup>',
        font_size=16,
        x=0,
    ),
    xaxis=dict(
        title='Average Happiness Score',
        range=[0, 8.5],   # zero baseline
        showgrid=True,
        gridcolor='#ebebeb',
    ),
    yaxis=dict(title=''),
    coloraxis_showscale=False,
    plot_bgcolor='white',
    paper_bgcolor='white',
    margin=dict(l=10, r=80, t=80, b=40),
    height=420,
)

fig.show()
```

## Task 2 — Bottom vs. Top: A Contrast Story

**What to build:** A bar chart that highlights the **gap between the happiest and least happy countries**, focusing on a specific insight.

**Requirements:**
- Show the **top 8 AND bottom 8 countries** together (16 bars total)
- Use **colour** to distinguish the two groups (not Plotly's default rainbow)
- Add a **visual separator or annotation** that emphasises the gap
- Insight title that tells the story of the gap

**Hint:** Use `pd.concat([df.nlargest(8,'Happiness_Score'), df.nsmallest(8,'Happiness_Score')])` to get both groups.

**Stretch goal:** Add a vertical reference line showing the global average.


```python
# Task 2: Top 8 vs. Bottom 8 contrast
# ------------------------------------

# Step 1: Get top and bottom countries
top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'
bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'

combined = pd.concat([bottom8, top8]).sort_values('Happiness_Score')
global_avg = df['Happiness_Score'].mean()
print(f"Global average: {global_avg:.2f}")

# Step 2: Build your chart
# YOUR CODE HERE

color_map = {'Top 8': '#1a9850', 'Bottom 8': '#d73027'}

fig = px.bar(
    combined,
    x='Happiness_Score',
    y='Country',
    orientation='h',
    color='Group',
    color_discrete_map=color_map,
    text=combined['Happiness_Score'].round(2),
)

fig.update_traces(textposition='outside', textfont_size=10)

# Global average reference line
fig.add_vline(
    x=global_avg,
    line_dash='dash',
    line_color='steelblue',
    line_width=1.5,
    annotation_text=f'Global avg {global_avg:.2f}',
    annotation_position='top',
    annotation_font_size=11,
    annotation_font_color='steelblue',
)

# Separator between the two groups
fig.add_hline(
    y=7.5,   # between index 7 (bottom8) and 8 (top8) after sort
    line_dash='dot',
    line_color='gray',
    line_width=1,
)

fig.update_layout(
    title=dict(
        text='A Happiness Gap of Nearly 6 Points Separates Finland from Afghanistan<br>'
             '<sup>The top 8 countries score more than twice as high as the bottom 8</sup>',
        font_size=16,
        x=0,
    ),
    xaxis=dict(
        title='Happiness Score',
        range=[0, 9],    # zero baseline
        showgrid=True,
        gridcolor='#ebebeb',
    ),
    yaxis=dict(title=''),
    plot_bgcolor='white',
    paper_bgcolor='white',
    legend=dict(title='', orientation='h', y=1.08, x=1, xanchor='right'),
    margin=dict(l=10, r=80, t=90, b=40),
    height=560,
)

fig.show()
```

## Done? Stretch Goal

If you finish both tasks with time to spare, try this:

**Task 3 (stretch):** Build a **grouped bar chart** comparing 2 sub-factors (e.g. `GDP_per_capita` and `Freedom`) across the 5 most populated regions. Use colour meaningfully and write an insight title.

Regions to include: `'Western Europe'`, `'Latin America'`, `'East Asia'`, `'Sub-Saharan Africa'`, `'South Asia'`


```python
# Stretch goal — grouped bar chart
# YOUR CODE HERE

regions_sel = ['Western Europe', 'Latin America and Caribbean',
               'East Asia', 'Sub-Saharan Africa', 'South Asia']

stretch = (df[df['Region'].isin(regions_sel)]
           .groupby('Region')[['GDP', 'Freedom']]
           .mean()
           .reset_index())

stretch_long = stretch.melt(
    id_vars='Region',
    value_vars=['GDP', 'Freedom'],
    var_name='Factor',
    value_name='Score'
)

# Shorten region name for readability
stretch_long['Region'] = stretch_long['Region'].str.replace(
    'Latin America and Caribbean', 'Latin America'
)

fig = px.bar(
    stretch_long,
    x='Region',
    y='Score',
    color='Factor',
    barmode='group',
    color_discrete_map={'GDP': '#4575b4', 'Freedom': '#fdae61'},
    text=stretch_long['Score'].round(2),
)

fig.update_traces(textposition='outside', textfont_size=10)

fig.update_layout(
    title=dict(
        text='Western Europe Leads in Both Wealth and Freedom; South Asia Lags on Both<br>'
             '<sup>GDP (log-scaled contribution) consistently outweighs perceived Freedom across all regions</sup>',
        font_size=15,
        x=0,
    ),
    xaxis=dict(title=''),
    yaxis=dict(
        title='Average Score',
        range=[0, 12],    # zero baseline
        showgrid=True,
        gridcolor='#ebebeb',
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    legend=dict(title='Factor', orientation='h', y=1.10, x=1, xanchor='right'),
    margin=dict(l=10, r=20, t=100, b=40),
    height=460,
)

fig.show()
```
