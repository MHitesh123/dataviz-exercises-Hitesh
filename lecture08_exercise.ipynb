## Task 1 — World choropleth: life expectancy diverging scale

**What to build:** A world choropleth showing **life expectancy relative to the global average** using a diverging colour scale.

**Requirements:**
- Use the Gapminder dataset for 2007: `px.data.gapminder()`
- Compute each country's deviation from the global mean life expectancy
- Diverging scale centred at zero (= world average)
- `hover_data` showing country name, raw life expectancy, and deviation
- Insight title naming which region is furthest below average

> 💡 `gm_2007['lifeExp'].mean()` gives you the global average to subtract from


````python
# Task 1
# YOUR CODE HERE

df = px.data.gapminder()
# Filter for 2007
df_2007 = df[df["year"] == 2007].copy()

global_mean = df_2007["lifeExp"].mean()

df_2007["lifeExp_dev"] = df_2007["lifeExp"] - global_mean

fig = px.choropleth(
    df_2007,
    locations="iso_alpha",  # 
    color="lifeExp_dev",
    hover_name="country",
    projection= "natural earth",
    hover_data={
        "lifeExp": True,
        "lifeExp_dev": True,
        "iso_alpha": False
    },
    color_continuous_scale="RdBu",  # Diverging scale
)

fig.update_coloraxes(cmid=0)

fig.update_layout(
    template="plotly_white",
    height=650,
    width=1000,
    title=dict(x=0.02, xanchor="left", font=dict(size=21)),
    font=dict(family="Arial", size=13, color="#1F2937"),
    margin=dict(l=20, r=20, t=90, b=20),
    coloraxis_colorbar=dict(
        
        ticksuffix=" yrs",
        
        
    )
)

fig.update_geos(
    showframe=False,
    showcoastlines=True,
    showcountries=True,
    fitbounds="locations",
    projection_scale=1.2
)

fig.update_layout(
    width=1000,
    height=800,
    title_text="Life Expectancy Deviation from Global Average (2007)"
)

fig.update_coloraxes(
    colorbar_title="Deviation (yrs)"

)
fig.show()
fig.write_html("choropleth_map_1.html")
````

````python
# Task 2
# YOUR CODE HERE

import plotly.express as px
import pandas as pd
import requests

# -----------------------------
# Load GeoJSON (US States)
# Source: https://github.com/PublicaMundi/MappingAPI
# -----------------------------
geojson_url = "https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json"
geojson = requests.get(geojson_url).json()

# Inspect properties (IMPORTANT STEP)
print(geojson['features'][0]['properties'])

# -----------------------------
# Create dataset (example: population per state)
# -----------------------------
data = {
    "state": ["Alabama","Alaska","Arizona","Arkansas","California","Colorado","Connecticut","Delaware","Florida","Georgia"],
    "value": [5.1, 0.7, 7.3, 3.0, 39.2, 5.8, 3.6, 1.0, 21.5, 10.7]
}

df = pd.DataFrame(data)

# -----------------------------
# Build choropleth
# -----------------------------
fig = px.choropleth(
    df,
    geojson=geojson,
    locations="state",
    featureidkey="properties.name",  # MATCHES GeoJSON
    color="value",
    color_continuous_scale="Viridis",  # Sequential scale
)

# Fit map to locations
fig.update_geos(fitbounds="locations", visible=False)

# Insight title
fig.update_layout(
    title="California and Florida Lead Population Concentration Among Selected US States"
)

fig.show()
fig.write_html("choropleth_map_2.html")
````
